```python
time = []
for sub in test.index.get_level_values(0).unique() : 
    time.extend(np.arange(len(test.loc[sub, : ])))
    #test.loc[sub, : ]["time"] 

test["Time"] = time
```

```python
### time step : 날짜와 정보의 선형회귀 분석
### Time-step features let you model time dependence
```

```python
for sub in test.index.get_level_values(0).unique() : 
    cond = test.loc[sub, :].sum(axis = 0) > 0
    #print(cond.sort_values()[-2:].index)
    target_col = cond.sort_values()[-2:].index
    #print(test.loc[sub, :][target_col].dropna())
    target_df = pd.DataFrame(test.loc[sub, :][target_col].dropna())
    #print(target_df[target_col[1]], target_df[target_col[0]])
    
    #plt.style.use("seaborn-whitegrid")
    plt.rc(
        "figure",
        autolayout=True,
        figsize=(11, 4),
        titlesize=18,
        titleweight='bold',
)
    plt.rc(
        "axes",
        labelweight="bold",
        labelsize="large",
        titleweight="bold",
        titlesize=16,
        titlepad=10,
)

    fig, ax = plt.subplots()
    ax.plot( target_col[1], target_col[0], data=target_df, color='0.75')
    ax = sns.regplot(x='Time', y= target_col[0], data=target_df, ci=None, scatter_kws=dict(color='0.25'))
    ax.set_title(target_col[0]);
   
```

```python
### lag feature we shift the observations of the target series
### lag features let you model serial dependence
```

```python
for sub in test.index.get_level_values(0).unique() : 
    cond = test.loc[sub, :].sum(axis = 0) > 0
    #print(cond.sort_values()[-2:].index)
    target_col = cond.sort_values()[-2:].index
    #print(test.loc[sub, :][target_col].dropna())
    target_df = pd.DataFrame(test.loc[sub, :][target_col].dropna())
    #print(target_df[target_col[0]], target_df[target_col[0]].shift())
    
    plt.style.use("seaborn-whitegrid")
    plt.rc(
        "figure",
        autolayout=True,
        figsize=(11, 4),
        titlesize=18,
        titleweight='bold',
)
    plt.rc(
        "axes",
        labelweight="bold",
        labelsize="large",
        titleweight="bold",
        titlesize=16,
        titlepad=10,
)

    fig, ax = plt.subplots()
    ax = sns.regplot(x=target_df[target_col[0]].shift(), y=target_df[target_col[0]], data=target_df, ci=None, scatter_kws=dict(color='0.25'))
    ax.set_aspect('equal')
    ax.set_title(f'Lag Plot of {target_col[0]}');
   
```
