```python
def bar_plot(variable) : 
    """
    input : df["Categorical Variable"]
    output : bar plot & value count
    """
    
    # get feature
    var = df[variable]
    # count num of categorical variable 
    value = var.value_counts()
    
    # visualize 
    plt.figure(figsize = (12,6))
    plt.bar(value.index, value)
    plt.ylabel("counts value")
    plt.title(variable)
    plt.xticks(rotation=45)
    plt.show()
    print(f"{variable} : \n{value}")
```

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(5, 10))

# Plot the transaction_amount
sns.set_color_codes("pastel")
sns.barplot(x="transaction_amount", y="category_title", data=amount,
            label="transaction_amount", color="b")

# Plot the sale_price
sns.set_color_codes("muted")
sns.barplot(x="sale_price", y="category_title", data=amount,
            label="sale_price", color="b")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 1500000), ylabel="",
       xlabel="sales info for each category sector")
sns.despine(left=True, bottom=True)
```
